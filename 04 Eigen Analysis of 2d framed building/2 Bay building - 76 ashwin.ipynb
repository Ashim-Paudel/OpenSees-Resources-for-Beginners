## Model Variables

- For time period analysis we will calculate section dimensions based on given data for stiffness of each column in story

```python
### GEOMETRIC DATA ###
# storey height: 
x_coors = np.array([0.0, 20*ft, 40*ft])
y_coors = np.array([0.0, 20*ft, 10*ft])


ops.wipe()
ops.model('basic', '-ndm', 2, '-ndf', 3)
#defining nodes

ops.node(1, 0.0, 0.0)
ops.node(2, 20*ft, 0)
ops.node(3, 40*ft, 0)
ops.node(4, 0*ft, 20*ft)
ops.node(5, 20*ft, 20*ft)
ops.node(6, 40*ft, 20*ft)
ops.node(7, 20*ft, 30*ft)
ops.node(8, 40*ft, 30*ft)

ops.fix(1 , *[1,1,0])
ops.fix(2, *[1,1,1])
ops.fix(3, *[1,1,0])

ops.equalDOF(8, 7, *[2,3])
ops.equalDOF(6, 5, *[2,3])
ops.equalDOF(5, 4, *[2,3])

ops.mass(4, *[40*kN/g, 0.0, 0.0])
ops.mass(5, *[80*kN/g, 0.0, 0.0])
ops.mass(6, *[40*kN/g, 0.0, 0.0])

ops.mass(7, *[50*kN/g, 0.0, 0.0])
ops.mass(8, *[50*kN/g, 0.0, 0.0])


ovs.plot_model()
```

```python
linTrans = 1
ops.geomTransf('Linear', linTrans)

EI = 120*kN*inch**2
E = 2e11*N/m**2
I = EI/E
Ib = 1000*I


b_col = (I*12)**(1/4)
A_col = b_col **2
A_beam = A_col


ops.element('elasticBeamColumn', 1, *[1, 4], A_col, E, I, linTrans)
ops.element('elasticBeamColumn', 2, *[2, 5], A_col, E, I, linTrans)
ops.element('elasticBeamColumn', 3, *[3, 6], A_col, E, I, linTrans)
ops.element('elasticBeamColumn', 4, *[5, 7], A_col, E, I, linTrans)
ops.element('elasticBeamColumn', 5, *[6, 8], A_col, E, I, linTrans)

ops.element('elasticBeamColumn', 6, *[4, 5], A_beam, E, Ib, linTrans)
ops.element('elasticBeamColumn', 7, *[5, 6], A_beam, E, Ib, linTrans)
ops.element('elasticBeamColumn', 8, *[7, 8], A_beam, E, Ib, linTrans)


ovs.plot_model()
```

#### For the specified number of eigenvalues (numModes) (for this example it is 3) the eigenvectors are recorded at all nodes in all DOFs using node recorder command

```python
numModes = 2

lamda = ops.eigen(numModes)

omegas = []
freqs = []
T = []
for lam in lamda:
    omega = np.sqrt(lam)
    freq  = omega/(2*np.pi)
    omegas.append(omega)
    freqs.append(freq)
    T.append(1/freq)
```

```python
print(f"The time period at 3 modes are: {T}")
print(f"The Fundamental period is: {max(T)}")
print(f"The Fundamental frequency is: {min(freqs)}")
```
