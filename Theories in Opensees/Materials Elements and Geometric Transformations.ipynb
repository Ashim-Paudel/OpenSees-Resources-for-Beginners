# Material Models
1. Constitutive Relationship $\left( \sigma = E \epsilon \right)$
2. Axial Force-Deformation and Moment-Rotation $\left( F = k x \right), \left( M = k \theta \right)$

# Types of Material Models

## Uniaxial Material
- One Dimensional Stress-Strain or Force-Deformation relationship.
  - Stress - Strain (fibre)
  - Moment - Curvature (Uniaxial Section Bending DOF)
  - Axial Force - Strain (Uniaxial Section axial DOF)
  - Moment - Rotation ( Rotational Spring)
  - Force - Deformation (Translational Spring)

## ND Material
- N-Dimensional Stress-Strain Relationship in a continuum element (2D or 3D)
- Eg. Soil modeling, modeling dams and so on

## Section
- Can be used in some beam-column elements to represent an effective moment-rotation behaviour at sample points
