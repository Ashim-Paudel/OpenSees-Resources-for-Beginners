````python
#setting global constants for geometry

span = 6 #in m
storey = 3.5

#main vertical grid lines
x1 = 0
x2 = x1+span
x = [x1, x2]

#horizontal grid lines
z0 = 0
z1 = z0+storey
z2 = z0+2*storey
z = [z0, z1, z2]
````

### nodes definition


</img>

- main nodes = 1,2,3,4,5,6
- extra nodes for connecting plastic hinges = 7,8,9,10

 - we will model **plastic hinges** as **two nodes located in same joint: connected by rotational zero length spring**  

````python
# setting names for node tags for main nodes
N_a0, N_b0, N_a1, N_b1, N_a2, N_b2 = 1,2,3,4,5,6

#for nodes for plastic hinges:
N_a1_ph, N_b1_ph, N_a2_ph, N_b2_ph = 7,8,9,10
````

````python
#nodes assignment for main nodes
for j in range(len(z)):
    for i in range(len(x)):
        k = len(x) * j + i
        ops.node(nodes[k], x[i], z[j])
        print(nodes[k], x[i], z[j])
````

````python
#nodes assignment for plastic hinges
for j in range(1, len(z)):
    for i in range(len(x)):
        k = len(x) * j + i
        print(k,nodes_ph[k-2], x[i], z[j])
        ops.node(nodes_ph[k-2], x[i], z[j])
````

### Restraints and Constraints defining:
- defining of support condition at each nodes

- **fix** is used for restraints at the supports for the structure
- **constraints** are ties between certain DoF of two different nodes. *equalDOF* is used.

````python
ops.fix(N_a0, 1, 1, 1) #restrained in all directions
ops.fix(N_b0, 1, 1, 1) #restrained in x, y and rotation
````

````python
#ops.equalDOF(N_a1, N_a1_ph, 1, 2) for all plastic hinges, we can achieve this in loops

for i in range(4):
    print(nodes[i+2], nodes_ph[i])
    ops.equalDOF(nodes[i+2], nodes_ph[i], 1, 2)  #equalDOF node1 node2 dof1(x) dof2(y) dof3(rotation)
````

### Sections Properties for Column elements

#### defining a function to create fibre section

**Section Command**
- `section secType? secTag? arg1? ...`
- **link** https://opensees.berkeley.edu/wiki/index.php/Section_Command
- **Fiber Section**
    - **link** https://opensees.berkeley.edu/wiki/index.php/Fiber_Section
    - ``` 
    
        section Fiber $secTag <-GJ $GJ> {
        fiber...
        patch...
        layer...
        ...
        }
        ```

(**patch command**)
- The patch command is used to generate a number of fibers over a cross-sectional area. Currently there are three types of cross-section that fibers can be generated: quadrilateral, rectangular and circular.
- https://opensees.berkeley.edu/wiki/index.php/Patch_Command (**patch command link**)
- `patch quad $matTag $numSubdivIJ $numSubdivJK $yI $zI $yJ $zJ $yK $zK $yL $zL`

- The figure shows the no. of fibres created in the section

<div></img>
</img>
</div>



````python
def Wsection(secID, matID, d, bf, tf, tw, nfdw, nftw, nfbf, nftf):
    # ###################################################################
    # Wsection  $secID $matID $d $bf $tf $tw $nfdw $nftw $nfbf $nftf
    # ###################################################################
    # create a standard W section given the nominal section properties
    # written: Remo M. de Souza
    # date: 06/99
    # modified: 08/99  (according to the new general modelbuilder)
    # input parameters
    # secID - section ID number
    # matID - material ID number 
    # d  = nominal depth
    # tw = web thickness
    # bf = flange width
    # tf = flange thickness
    # nfdw = number of fibers along web depth 
    # nftw = number of fibers along web thickness
    # nfbf = number of fibers along flange width
    # nftf = number of fibers along flange thickness
    
    dw= d - 2 * tf
    y1= -d/2
    y2= -dw/2
    y3= dw/2
    y4= d/2
    
    z1= -bf/2
    z2= -tw/2
    z3= tw/2
    z4= bf/2
    
    ops.section('fiberSec', secID)
    ops.patch('quadr', matID, nfbf, nftf, y1, z4, y1, z1, y2, z1, y2, z4)
    ops.patch('quadr', matID, nftw, nfdw, y2, z3, y2, z2, y3, z2, y3, z3)
    ops.patch('quadr', matID, nfbf, nftf, y3, z4, y3, z1, y4, z1, y4, z4)
        #                         nfIJ, nfJK, yI, zI, yJ, zJ, yK, zK, yL, zL

                
    #patch quad $matTag $numSubdivIJ $numSubdivJK $yI $zI $yJ $zJ $yK $zK $yL $zL
    #$matTag	tag of previously defined material (UniaxialMaterial tag for a FiberSection or NDMaterial tag for use in an NDFiberSection)
    #$numSubdivIJ	number of subdivisions (fibers) in the IJ direction.
    #$numSubdivJK	number of subdivisions (fibers) in the JK direction.
    #$yI $zI	y & z-coordinates of vertex I (local coordinate system)
    #$yJ $zJ	y & z-coordinates of vertex J (local coordinate system)
    #$yK $zK	y & z-coordinates of vertex K (local coordinate system)
    #$yL $zL	y & z-coordinates of vertex L (local coordinate system)
````

````python
# section id:
HE180B = 1

#column dimensions
hc1 = 0.1800 #depth of section
bc1 = 0.1800 #width of section
tfc1 = 0.0140 #flange thickness
twc1 = 0.0085 #web thickness

#now define the section with function and other items created with fiber section using the function
Wsection(HE180B, s355, hc1, bc1, tfc1, twc1, 4, 2, 4, 2)
````

### Plotting the generated fiber section



````python
import matplotlib.pyplot as plt


d = hc1 * 10 #scale 100 times
bf = bc1 * 10
tf = tfc1 * 10
tw = twc1 * 10

dw= d - 2 * tf
y1= -d/2
y2= -dw/2
y3= dw/2
y4= d/2

z1= -bf/2
z2= -tw/2
z3= tw/2
z4= bf/2
    
nfdw, nftw, nfbf, nftf = 4, 2, 4, 2
fib_sec_1 = [['section', 'Fiber', HE180B],
             ['patch', 'quad', s355, nfbf, nftf, y1, z4, y1, z1, y2, z1, y2, z4],
             ['patch', 'quad', s355, nftw, nfdw, y2, z3, y2, z2, y3, z2, y3, z3],
             ['patch', 'quad', s355, nfbf, nftf, y3, z4, y3, z1, y4, z1, y4, z4],
             ]

matcolor = ['r', 'lightgrey', 'gold', 'w', 'w', 'w']
ovs.plot_fiber_section(fib_sec_1, matcolor=matcolor)
plt.axis([-1, 1, -1, 1])
plt.title("Fiber Section of the Column Element")
````

````python
with open("Outputs/ModalAnalysis/NodeEigenVectos_EigenValues.txt", 'w') as eigenFile:
    lamdas = ops.eigen(2)
    eigenFile.write("lambda\t omgea\t period\t frequency\n")
    for lamda in lamdas:
        omega = np.sqrt(lamda)
        period = 2*np.pi/omega
        frequency = 1/period
        eigenFile.write(f"{lamda:.2f}\t {omega:.2f}\t {period:.2f}\t {frequency:.2f}\n")

ops.record()
````

````python
#getting eigenvalue1 for the 1st and 2nd storeys

eig_val = np.loadtxt('Outputs/ModalAnalysis/NodeEigenVectors_EigenVal1.txt', delimiter=" ")
````

````python
eig_val
#time #storey0 #storey1 #storey2
````

````python
phi1 = abs(eig_val[2])
phi2 = abs(eig_val[3])
````

````python
#load defining
horizontalLoadPattern = 2
ops.pattern('Plain', horizontalLoadPattern, timeSeriesTag)

# Apply lateral load based on first mode shape in x direction (EC8-1)
ops.load(N_a1, *[phi1*mass1, 0.0, 0.0])
ops.load(N_a2, *[phi2*mass2, 0.0, 0.0])
ops.load(N_b1, *[phi1*mass1, 0.0, 0.0])
ops.load(N_b2, *[phi2*mass2, 0.0, 0.0])
````

````python
step = 1.000e-4
numSteps = 5000

#analysis commands
ops.constraints("Transformation")
ops.numberer("RCM")
ops.system("BandGeneral")
ops.test("NormDispIncr", 0.000001, 100)
#algorithm('NewtonLineSearch', Bisection=False, Secant=False, RegulaFalsi=False, InitialInterpolated=False, tol=0.8, maxIter=10, minEta=0.1, maxEta=10.0)
ops.algorithm("NewtonLineSearch", '-type', 'Bisection', '-tol', 8e-1,'maxIter',1000,  '-minEta', 1e-1, '-maxEta', 1e1, 'pFLag', 1)
#integrator('DisplacementControl', nodeTag, dof, incr, numIter=1, dUmin=incr, dUmax=incr)
ops.integrator('DisplacementControl', N_a2, 1, step)
ops.analysis("Static")

ops.record()
t1 = time.time()
ops.analyze(numSteps)
t2 = time.time()
print(f"Time Taken for analysis = {t2-t1} sec.")
````
